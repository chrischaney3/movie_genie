```python
import pandas as pd
import numpy as np
import gzip
from IPython.display import display
```

```python
df_credits = pd.read_csv('datasets/credits.csv')
```

```python
df_credits
```

```python
# Convert elements in 'cast' and 'crew' columns to strings
df_credits['cast'] = df_credits['cast'].astype(str)
df_credits['crew'] = df_credits['crew'].astype(str)

# Apply eval() to evaluate the string representation of each element
df_credits['cast'] = df_credits['cast'].apply(eval)
df_credits['crew'] = df_credits['crew'].apply(eval)

# Create new columns to store the extracted information
df_credits['cast_names'] = ""
df_credits['cast_characters'] = ""
df_credits['cast_credit_ids'] = ""
df_credits['cast_profile_paths'] = ""
df_credits['crew_director_names'] = ""
df_credits['crew_executive_producers'] = ""

# Iterate over each row and extract the desired information
for index, row in df_credits.iterrows():
    cast_characters = [cast_member['character'] for cast_member in row['cast']]
    cast_credit_ids = [cast_member['credit_id'] for cast_member in row['cast']]
    cast_profile_paths = [cast_member['profile_path'] for cast_member in row['cast']]
    
    crew_director_names = [crew_member['name'] for crew_member in row['crew'] if crew_member['job'] == 'Director']
    crew_executive_producers = [crew_member['name'] for crew_member in row['crew'] if crew_member['job'] == 'Executive Producer']
    
    df_credits.at[index, 'cast_characters'] = cast_characters
    df_credits.at[index, 'cast_credit_ids'] = cast_credit_ids
    df_credits.at[index, 'cast_profile_paths'] = cast_profile_paths
    df_credits.at[index, 'crew_director_names'] = crew_director_names
    df_credits.at[index, 'crew_executive_producers'] = crew_executive_producers

# Print the updated dataframe
df_credits
```

```python
# Drop columns 'B' and 'C'
df_credits = df_credits.drop(['cast', 'crew', 'id', 'cast_names'], axis=1)

# Print the updated DataFrame
df_credits
```

```python
# Define the output filename
output_filename = 'df_credits.csv.gz'

# Compress and save the dataframe
with gzip.open(output_filename, 'wt', compresslevel=5) as file:
    df_credits.to_csv(file, index=False)
```

```python
df_keywords = pd.read_csv('datasets/keywords.csv')
```

```python
df_keywords
```

```python
df_keywords = df_keywords[['keywords']]
df_keywords
```

```python
import re
# Create a new column to store the cleaned keywords
df_keywords['cleaned_keywords'] = ""

# Iterate over each row and clean the keywords
for index, row in df_keywords.iterrows():
    # Extract the keywords using regular expression pattern
    keywords = re.findall(r'\b\w+\b', str(row['keywords']))
    
    # Filter out any numbers or IDs
    keywords = [word for word in keywords if not any(char.isdigit() for char in word)]
    
    # Assign the cleaned keywords to the new column
    df_keywords.at[index, 'cleaned_keywords'] = ', '.join(keywords)

# Print the updated DataFrame
df_keywords
```

```python
import re


# Create a new column to store the cleaned keywords
df_keywords['cleaned_keywords'] = ""

# Iterate over each row and clean the keywords
for index, row in df_keywords.iterrows():
    # Extract the keywords using regular expression pattern
    keywords = re.findall(r'\b\w+\b', str(row['keywords']))
    
    # Filter out any numbers or IDs
    keywords = [word for word in keywords if not any(char.isdigit() for char in word)]
    
    # Remove specific words like 'id' and 'name'
    keywords = [word for word in keywords if word.lower() not in ['id', 'name']]
    
    # Assign the cleaned keywords to the new column
    df_keywords.at[index, 'cleaned_keywords'] = ', '.join(keywords)

# Select only the cleaned keywords column
cleaned_keywords_column = df_keywords['cleaned_keywords']

# Print the cleaned keywords column
cleaned_keywords_column
```

```python
# Define the output filename
output_filename = 'df_cleaned_keywords.csv.gz'

# Compress and save the dataframe
with gzip.open(output_filename, 'wt', compresslevel=5) as file:
    cleaned_keywords_column.to_csv(file, index=False)
```
