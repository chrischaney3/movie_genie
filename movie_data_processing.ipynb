```python
import pandas as pd
```

```python
metadata_df = pd.read_csv("movies_metadata.csv")
metadata_df.head(2)
```

```python
metadata_df.info()
```

```python
metadata_df = metadata_df.drop(["adult", "belongs_to_collection", "homepage", "original_title", "overview", "popularity", "production_countries", "spoken_languages", "tagline"], axis=1)
```

```python
metadata_df
```

```python
metadata_df = metadata_df.astype({"video": str}, errors="raise")
metadata_df
```

```python
metadata_df.info()
```

```python
metadata_df = metadata_df[metadata_df["video"] == "False"]
metadata_df
```

```python
metadata_df = metadata_df.drop(["status", "video"], axis=1)
metadata_df
```

```python
credits_df = pd.read_csv("credits.csv")
keywords_df = pd.read_csv("keywords.csv")
```

```python
credits_df.info()
```

```python
keywords_df.info()
```

```python
metadata_df = metadata_df.astype({"id": int}, errors="raise")
```

```python
movies_df = pd.merge(metadata_df, credits_df, on="id")
movies_df
```

```python
movies_df = pd.merge(movies_df, keywords_df, on="id")
```

```python
movies_df
```

```python
movies_df = movies_df.astype({"cast": str, "crew": str, "keywords": str}, errors="raise")
movies_df = movies_df.rename(columns={"keywords": "unextracted_keywords"})
movies_df = movies_df.rename(columns={"genres": "unextracted_genres"})
movies_df = movies_df.rename(columns={"production_companies": "unextracted_production_companies"})
```

```python
movies_df["cast_names"] = ""
movies_df["director_name"] = ""
movies_df["executive_producers"] = ""
movies_df["keywords"] = ""
movies_df["genres"] = ""
movies_df["production_companies"] = ""
```

```python
movies_df["cast"] = movies_df["cast"].apply(eval)
movies_df["crew"] = movies_df["crew"].apply(eval)
movies_df["unextracted_keywords"] = movies_df["unextracted_keywords"].apply(eval)
movies_df["unextracted_genres"] = movies_df["unextracted_genres"].apply(eval)
movies_df["unextracted_production_companies"] = movies_df["unextracted_production_companies"].apply(eval)
```

```python
for index, row in movies_df.iterrows():
    cast_names = [cast_member["name"] for cast_member in row["cast"]]
    director_names = [crew_member["name"] for crew_member in row["crew"] if crew_member["job"] == "Director"]
    producer_names = [crew_member["name"] for crew_member in row["crew"] if crew_member["job"] == "Executive Producer"]
    keywords = [keyword["name"] for keyword in row["unextracted_keywords"]]
    genres = [genre["name"] for genre in row["unextracted_genres"]]
    production_companies = [production_company["name"] for production_company in row["unextracted_production_companies"]]
    
    movies_df.at[index, "cast_names"] = cast_names
    movies_df.at[index, "director_name"] = director_names
    movies_df.at[index, "executive_producers"] = producer_names
    movies_df.at[index, "keywords"] = keywords
    movies_df.at[index, "genres"] = genres
    movies_df.at[index, "production_companies"] = production_companies
```

```python
movies_df
```

```python
movies_df = movies_df.drop(["cast", "crew", "unextracted_keywords", "unextracted_genres", "unextracted_production_companies"], axis=1)
movies_df.head()
```

```python
movies_df.to_csv("movies_cleaned.csv")
```
